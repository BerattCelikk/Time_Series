### Calculating the Residuals

Calculate the residuals by subtracting the predictions from the data

Plot the residuals


```python
residuals = np.array(data['CO2']) -\
trend_model.predict(np.array(data.index).reshape((-1,1)))
plt.figure(figsize=(5.5, 5.5))
pd.Series(data=residuals, index=data.index).plot(color='b')
plt.title('Residuals of trend model for CO2 concentrations')
plt.xlabel('Time')
plt.ylabel('CO2 concentration')
plt.xticks(rotation=30)
#plt.savefig('plots/Section 1/_01_07.png', format='png', dpi=300)
```

### Seasonality

Seasonality is manifested as periodic deviations such as those seen in the de-trended observations of CO2 emissions.

A practical technique of determining seasonality is through exploratory data analysis through the following plots:

- Run sequence plot
- Seasonal sub series plot
- Multiple box plots


```python
data
```

```python
# Add a new column 'Residuals' to the DataFrame 'data' containing the previously calculated residuals
data['Residuals'] = residuals

# Create a mapping of month numbers to their corresponding quarters
month_quarter_map = {
    1: 'Q1', 2: 'Q1', 3: 'Q1',   # January to March are in Q1
    4: 'Q2', 5: 'Q2', 6: 'Q2',   # April to June are in Q2
    7: 'Q3', 8: 'Q3', 9: 'Q3',   # July to September are in Q3
    10: 'Q4', 11: 'Q4', 12: 'Q4'  # October to December are in Q4
}

# Map each month in the 'Month' column to its corresponding quarter using a lambda function
data['Quarter'] = data['Month'].map(lambda m: month_quarter_map.get(m))
```

```python
# Grouping the data by 'Year' and 'Quarter' and aggregating the 'Residuals' column
seasonal_sub_series_data = data.groupby(
    by=['Year', 'Quarter']  # Group the data by 'Year' and 'Quarter'
)['Residuals'].aggregate([np.mean, np.std])  # Calculate the mean and standard deviation of 'Residuals' for each group
```

```python
seasonal_sub_series_data.columns =\
['Quarterly Mean', 'Quarterly Standard Deviation']
```

### Zero-mean Model or White Noise
Once predictions have been made by a time series forecast model, they can be collected and analyzed. The series of forecast errors should ideally be white noise. When forecast errors are white noise, it means that all of the signal information in the time series has been harnessed by the model in order to make predictions.

### Random Walk
A random walk is different from a list of random numbers because the next value in the sequence is a modification of the previous value in the sequence.

$$y(t)=B_0+B_1\times y(t-1) + e(t)$$

$e(t)$ is a random noise or increment at time t. It represents the random fluctuation or innovation added to the last observed value to obtain the next value in the time series.

Given the way that the random walk is constructed, we would expect a strong autocorrelation with the previous observation and a linear fall off from there with previous lag values.

```python
random_walk1 = np.cumsum(zero_mean_series)  # Calculate the cumulative sum of the zero_mean_series array
plt.figure(figsize=(5.5, 5.5))  # Create a new figure for the plot with a specified size
plt.plot(random_walk1)  # Plot the random walk data
plt.title('Random Walk')  # Set the title of the plot to 'Random Walk'
plt.xlabel('Time index')  # Label the x-axis as 'Time index'
# Uncomment the next line to save the figure as a PNG file with a specified DPI
# plt.savefig('plots/Section 1/_01_13.png', format='png', dpi=300)
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(random_walk1)
```

```python
# Create and plot a random walk

from random import seed  # Import the seed function from the random module to control random number generation
from random import random  # Import the random function to generate random float numbers in the range [0.0, 1.0)

seed(1)  # Set the seed for the random number generator to ensure that the output is reproducible
# By setting a seed, you can obtain the same sequence of random numbers each time you run the code,
# which is essential for debugging and validating results.

random_walk = list()  # Initialize an empty list to store the values of the random walk

# Start the random walk with an initial value of -1 or 1 based on a random decision
random_walk.append(-1 if random() < 0.5 else 1)  
# The random function generates a float between 0.0 and 1.0. If the generated value is less than 0.5,
# append -1 to the random_walk list; otherwise, append 1. This creates the first step of the random walk.

# Generate the next 999 steps of the random walk
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1  # Determine the direction of movement for the next step
    # Again, a random value decides whether to move -1 (down) or +1 (up).
    
    value = random_walk[i-1] + movement  # Calculate the new value by adding the movement to the previous value
    # This cumulative sum creates the random walk, where each step is determined by the previous position plus a random movement.

    random_walk.append(value)  # Append the new value to the random_walk list

plt.plot(random_walk)  # Plot the generated random walk
# This visualizes the trajectory of the random walk over the indices from 0 to 999.
# The resulting plot will show a jagged line that represents the random movements, 
# which can be used to analyze the characteristics of a random walk, such as trends, fluctuations, and volatility.
```

```python
djia_df = pd.read_excel('DJIA_Jan2016_Dec2016.xlsx')
djia_df.head(10)
```

```python
plt.figure(figsize=(5.5, 5.5))
djia_df['Close'].plot(color='b')
plt.title('Dow Jones Industrial Average between Jan 2016 - Dec 2016')
plt.xlabel('Time')
plt.ylabel('Closing Value')
#plt.savefig('plots/Section 1/_01_14.png', format='png', dpi=300)
```

```python
# Define a range of lag values from 0 to 30
lag = range(0, 31)

# Initialize an empty list to store the autocorrelation values for each lag
djia_acf = []

# Iterate over each lag value in the defined range
for l in lag:
    # Calculate the autocorrelation of the 'Close' column in the djia_df DataFrame for the current lag 'l'
    # Append the calculated autocorrelation value to the djia_acf list
    djia_acf.append(djia_df['Close'].autocorr(l))  
    # The autocorr method computes the correlation of the 'Close' price with its own lagged values.
    # This helps to understand how past values of the 'Close' price influence its current value.
```

```python
plt.figure(figsize=(5.5, 5.5))
plt.plot(djia_acf, marker='.', color='b')
plt.title('Autocorrelation function for DJIA')
plt.xlabel('Lag in terms of number of trading days')
plt.ylabel('Autocorrelation function')
#plt.savefig('plots/Section 1/_01_15.png', format='png', dpi=300)
```

```python
# Importing functions to plot autocorrelation and partial autocorrelation functions
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
# 'plot_acf' is used to visualize the autocorrelation of a time series, showing how the series is correlated with its past values.
# It helps in identifying patterns like seasonality and trends.
# 'plot_pacf' is used to visualize the partial autocorrelation of a time series, which measures the correlation between a time series and its lagged values after removing the effects of intermediate lags.
# This is particularly useful in determining the appropriate order of an AR (AutoRegressive) model for time series forecasting.
```

```python
# Plot autocorrelation and confidence intervals using the plot_acf function
plt.figure(figsize=(5.5, 5.5))  # Create a new figure with a specified size for better visualization
plot_acf(djia_df['Close'], lags=20)  # Plot the autocorrelation function for the 'Close' prices in the DJIA DataFrame, showing correlations up to 20 lags
#plt.savefig('plots/Section 1/_01_16.png', format='png', dpi=300)  # Uncomment this line to save the plot as a PNG file with specified resolution
```

```python
# Plot partial autocorrelation and confidence intervals using the plot_pacf function
plt.figure(figsize=(5.5, 5.5))  # Create a new figure with a specified size for clear visualization
plot_pacf(djia_df['Close'], lags=20)  # Plot the partial autocorrelation function for the 'Close' prices in the DJIA DataFrame, showing correlations up to 20 lags
#plt.savefig('plots/Section 1/_01_17.png', format='png', dpi=300)  # Uncomment this line to save the plot as a PNG file with specified resolution
```
